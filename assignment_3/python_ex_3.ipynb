```python
# import pandas library
import pandas as pd
```

Read the CSV

```python
kiva_data = pd.read_csv('./kiva_loans.csv')
print(f'Number of entries: {kiva_data.shape[0]:,}\n'
      f'Number of features: {kiva_data.shape[1]:,}\n\n')
```

```python
kiva_data.head()
```

```python
#get list of countries
country_list = kiva_data['country'].unique().tolist()

country_list
```

```python
#get list of activities
activity_list = kiva_data['activity'].unique().tolist()

activity_list
```

```python
# Select rows where the country is either 'Kenya' or 'El Salvador'
kenya_or_elsalvador = kiva_data[kiva_data.country.isin(['Kenya', 'El Salvador'])]
print(f'{kenya_or_elsalvador.shape[0]:,} rows where country is Kenya or El Salvador')

# Select rows where the loan amount is greater than 3000
loan_amount_gt_1000 = kiva_data[kiva_data.loan_amount > 3000]
print(f'{loan_amount_gt_1000.shape[0]:,} rows where loan amount is greater than 3000')

# Select rows where the country is 'Kenya' and the loan amount is greater than 3000
pakistan_and_loan_amount_gt_1000 = kiva_data[(kiva_data.country == 'Pakistan') & (kiva_data.loan_amount > 3000)]
print(f'{pakistan_and_loan_amount_gt_1000.shape[0]:,} rows where country is Kenya and loan amount is greater than 3000')
```

Sort the table by activity:

```python
# Sort table using one or more columns
kiva_data.sort_values(by='activity', ascending=True)
```

Look at the mean loan amount per activity:

```python
# Group and aggregate rows with "groupby"
kiva_data.groupby('activity').agg({'loan_amount': 'mean'})
```

And by country:

```python
# Group and aggregate rows with "groupby"
kiva_data.groupby('country').agg({'loan_amount': 'mean'})
```

Use apply by:

```python
# Create a new variable using apply and lambda function
kiva_data['term_in_years'] = kiva_data['term_in_months'].apply(lambda x: x/12)
kiva_data.head()
```

Find loan to value percent per activity

```python
# Calculate the mean of funded_amount for each activity
funded_amount_mean_per_activity = kiva_data.groupby('activity')['funded_amount'].mean()

# Calculate the mean of loan_amount for each activity
loan_amount_mean_per_activity = kiva_data.groupby('activity')['loan_amount'].mean()

# Subtract the mean funded_amount from the mean loan_amount for each activity
loan_to_value_per_activity = funded_amount_mean_per_activity/loan_amount_mean_per_activity*100

loan_to_value_per_activity = loan_to_value_per_activity.rename('loan_to_value %')
loan_to_value_per_activity
```
